## Description

This notebook defines the 4D reconstruction kernels via 1D Kernel Density Estimation (KDE) of the reconstructed histograms, which will eventually be used by the `RecoServiceKernelFile` service in the Reco stage of PISA.

It has been seen that simply giving the `RecoServiceKDE` service the distribution of reconstructed events in the true energy bin and letting it automatically determine the KDE that best smooths the reconstruction kernel is not adequate, because there are too many badly reconstructed events which skews the distribution, making it much too broad. To get around this issue, we need to "clean" the reconstruction events before giving it to the KDE, and we also will want to check to see which is the best KDE smoothing function to use. This is best done (checked) manually and must be done for each binning in the analysis.

### Function Definitions

```python
# Define all functions to use later:

def get_arb_cuts(data,cut_list,cut_downgoing=False):
    '''
    Make arbitrary set of cuts, defined from cut_list and data, which is a
    PyTables file object.
    '''
    conditions = []
    try:
        conditions = [data.__getattr__(cut[0]).col(cut[1]) == cut[2] for cut in cut_list]
    except:
        conditions = [data.__getattribute__(cut[0]).col(cut[1]) == cut[2] for cut in cut_list]

    if cut_downgoing:
        conditions.append(np.cos(data.MCNeutrino.col('zenith'))<0.)
        
    return np.alltrue(np.array(conditions),axis=0)

        
def get_resolution_data(data,cuts,reco_string='MultiNest_8D_Neutrino'):
    '''
    Forms arrays of unweighted true/reco energy/coszen from the data_files
    '''

    logging.warn('Getting reconstructions from: %s'%reco_string)
    
    true_egy = data.MCNeutrino.col('energy')[cuts]
    true_cz = np.cos(data.MCNeutrino.col('zenith'))[cuts]
    
    try:
        reco_cz = np.cos(data.__getattr__(reco_string).col('zenith'))[cuts]
        reco_egy = data.__getattr__(reco_string).col('energy')[cuts]
    except:
        reco_cz = np.cos(data.__getattribute__(reco_string).col('zenith'))[cuts]
        reco_egy = data.__getattribute__(reco_string).col('energy')[cuts]

    arrays = [true_egy,true_cz,reco_egy,reco_cz]

    return arrays
    
    
def check_bin_events(in_bin,width,ie,energy,bins,e_true,
                         min_events = 300,min_width=0.5):                  
    '''
    Checks whether there are enough events inside the bin to accurately 
    characterize the KDE from the reco data. If not, then it extends the 
    bin width to ensure an adequate number of events.
    '''
    nevents = np.sum(in_bin)
    lo_indx = ie
    hi_indx = ie+1
    min_edge = bins[lo_indx]
    max_edge = bins[hi_indx]
    if ((nevents < min_events) or (width < min_width) ):                            
                                                                              
        while( (nevents < min_events) or (width < min_width) ):     
            # Decrement lower bin edge if not at lowest:                                    
            if lo_indx > 0: lo_indx -= 1                                                    
            min_edge = bins[lo_indx]                                                  
                                                                                                    
            # Try increasing upper bin edge if not at max bin:                              
            hi_indx += 1                                                                    
            try:                                                                            
                max_edge = bins[hi_indx]                                              
            except:                                                                         
                hi_indx -= 1                                                                
                max_edge = bins[hi_indx]                                              
                                                                                                    
            width = (max_edge - min_edge)                                               
            in_bin = np.alltrue(np.array([e_true >= min_edge,                               
                                          e_true < max_edge,
                                          np.fabs(e_true - e_reco) <= e_true,
                                          np.fabs(cz_true - cz_reco) <= 1.2
                                          ]),axis=0)                       
            nevents = np.sum(in_bin)
            
    #print("    >>Using %d bins for nevents: %d and width: %.2f"
    #      %((hi_indx - lo_indx),nevents,width))
    return in_bin,min_edge,max_edge
```

### Now Construct the KDEs and compare to raw distribution by inspection

```python
#//////////////////////////////////////////////////////////
# This is what you change for each flavour interactively
#/////////////////////////////////////////////////////////
data_cur = nutau_data.root
flav = 'nutau'
int_type = 1
#//////////////////////////////////////////////////////////


cutlist = [('Cuts_V5_Step1','value',True),
           ('Cuts_V5_Step2','value',True),
           ('I3MCWeightDict','InteractionType',int_type)]

# Here, we want to get both nu/nubar and only cc events
cuts = get_arb_cuts(data_cur,cutlist,cut_downgoing=True)

e_true,cz_true,e_reco,cz_reco = get_resolution_data(data_cur,cuts)
print "total events: ",len(e_reco)
```

```python
#print "ebins: ",ebins
#print "ecen: ",ecen

nbins = 41
print "nbins: ",nbins

kde_dict[flav] = {'energy':[],'coszen':[]}

# Loop over all energies and plot KDE estimate of energy and coszen:
for iegy, energy in enumerate(ecen):

    emin = ebins[iegy]
    emax = ebins[iegy+1]
    bin_width = (emax - emin)
    
    #NOTE: the extra condition np.fabs(e_reco-e_true) to avoid
    # VERY poorly mis-reconstructed events, that we won't be
    # fitting anyway
    
    in_bin = np.alltrue(np.array([e_true >= emin,e_true<emax,
                                  np.fabs(e_reco-e_true) <= e_true
                                  ]),axis=0)
    
    #///////////////////////////////////
    # First do energy KDE and PDF
    plt.figure(figsize=(8,6))
    nevents = np.sum(in_bin)
    print "energy: %.2f GeV, nevents: %.2f"%(energy,nevents)
    
    in_bin,emin,emax = check_bin_events(in_bin,bin_width,iegy,energy,ebins,e_true,min_events = 300)
    nevents = np.sum(in_bin)
    logging.info("AFTER correcting: nevents=%.2f emin=%.2f, emax=%.2f"%(nevents,emin,emax))
    #print "     correcting): ",nevents
    
    e_res_data = e_reco[in_bin] - e_true[in_bin]
    
    # Bin the events into a hisogram to compare the KDE to:
    # configure binning:
    if flav in ['nutau','nc']:
        logging.warn('binning will be done for nutau,nc -> expecting offset from true Egy.')
        max_bin = emax*1.3
        bins_fit = np.linspace(-max_bin,emax,nbins)
    else:
        max_bin = energy*1.3
        bins_fit = np.linspace(-max_bin,max_bin,nbins)

    histo,bins,_ = plt.hist(e_res_data,bins=bins_fit,normed=True,
                            histtype='step',lw=2,label='cleaned distribution',alpha=0.7)
    
    in_bin_raw = np.alltrue(np.array([e_true >= emin, e_true<emax,]),axis=0)
    e_res_data_raw = e_reco[in_bin_raw] - e_true[in_bin_raw]
    histo,bins,_ = plt.hist(e_res_data_raw,bins=bins_fit,normed=True,
                            histtype='step',lw=2,label='raw distribution',color='r')
    
    bincenters = get_bin_centers(bins_fit)
    
    # Now make the kde and evaluate at bin centers:
    kde_egy = gaussian_kde(e_res_data,bw_method='silverman')
    pdf_egy = kde_egy.evaluate(bincenters)
    print "    >>computed kde_egy.factor: ",kde_egy.factor
        
    plt.plot(bincenters,pdf_egy,color='green',alpha=0.7,lw=3,label='KDE')
    
    plt.xlabel(r'$E_\mathrm{reco} - E_\mathrm{true}$',fontsize='xx-large')
    plt.title('Energy resolution for %s at Energy of %.2f GeV'%(flav,energy))
    plt.grid()
    plt.legend(loc='upper right')
    if save_plots:
        fname_string = '%s_energy_kde_%02.2fGeV.png'%(flav,energy) if energy >= 10.0 else '%s_energy_kde_0%.2fGeV.png'%(flav,energy)
        outfile = os.path.join(plot_dir,flav,fname_string)
        logging.warn("SAVING to file: %s"%outfile)
        plt.savefig(outfile,dpi=160)
    #/////////////////////////////////////////////////////////////
    
    #//////////////////////////////////////////
    # Now do coszen KDE and PDF
    plt.figure(figsize=(8,6))
    nevents = np.sum(in_bin)
    bins_fit = np.linspace(-1.0,1.2,nbins)
        
    cz_res_data = cz_reco[in_bin] - cz_true[in_bin]
    print "max: ",np.max(cz_res_data)
    histo,bins,_ = plt.hist(cz_res_data,bins=bins_fit,normed=True,
                            histtype='step',lw=2,label='cleaned distribution',alpha=0.7)
    
    in_bin_raw = np.alltrue(np.array([e_true >= emin, e_true<emax,]),axis=0)
    cz_res_data_raw = cz_reco[in_bin_raw] - cz_true[in_bin_raw]
    histo,bins,_ = plt.hist(cz_res_data_raw,bins=bins_fit,normed=True,
                            histtype='step',lw=2,label='raw distribution',color='r')
    
    bincenters = get_bin_centers(bins_fit)
    
    # Now make the kde and evaluate at bin centers:
    kde_cz = gaussian_kde(cz_res_data,bw_method='silverman')
    pdf_cz = kde_cz.evaluate(bincenters)
    print "    >>computed kde_cz.factor: ",kde_cz.factor
        
    plt.plot(bincenters,pdf_cz,color='green',alpha=0.7,lw=3,label='KDE')
    
    plt.xlabel(r'$(\cos\theta_z)_\mathrm{reco} - (\cos\theta_z)_\mathrm{true}$',fontsize='xx-large')
    plt.title('cos zenith resolution for %s at Energy of %.2f GeV'%(flav,energy))
    plt.grid()
    plt.legend(loc='upper right')
    if save_plots:
        fname_string = '%s_coszen_kde_%.2fGeV.png'%(flav,energy) if energy >= 10.0 else '%s_coszen_kde_0%.2fGeV.png'%(flav,energy)
        outfile = os.path.join(plot_dir,flav,fname_string)
        #outfile = os.path.join(plot_dir,flav,'%s_coszen_kde_%02.2fGeV.png'%(flav,energy))
        logging.warn("SAVING to file: %s"%outfile)
        plt.savefig(outfile,dpi=160)
    #/////////////////////////////////////////////////////////////
    
    
    kde_dict[flav]['energy'].append(kde_egy)
    kde_dict[flav]['coszen'].append(kde_cz)
    
    plt.show()
    #plt.clf()
```

## When ALL Flavours of `kde_dict` are defined, convert format of `kde_dict` to format used by PISA, and give to RecoServiceKDE.

```python
kde_dict_pisa = {flav: {'cc': {'energy':None,'coszen':None}, 'nc': {'energy':None,'coszen':None}} for flav in ['nue','numu','nutau']}
print kde_dict_pisa.items()
#print "kde_dict: "
#print kde_dict.items()

for key in kde_dict_pisa.keys():
    
#    print kde_dict[key]['energy']
    
    kde_dict_pisa[key]['cc']['energy'] = kde_dict[key]['energy']
    kde_dict_pisa[key]['cc']['coszen'] = kde_dict[key]['coszen']
    
    kde_dict_pisa[key]['nc']['energy'] = kde_dict['nc']['energy']
    kde_dict_pisa[key]['nc']['coszen'] = kde_dict['nc']['coszen']

print "kde_dict_pisa computed: "
#print kde_dict_pisa.items()
```

Now use `RecoServiceKDE` to calculate the kernels and save as `kernelfile`

```python
# Now use RecoServiceKDE to calculate kernels and save for use in PISA
reco_service = RecoServiceKDE(ebins,czbins)

kernels = reco_service.calculate_kernels(kde_dict=kde_dict_pisa)
reco_service.kernels = kernels
#print "  keys: ",kernels.keys()

outfile = os.path.join(outdir,'V36_kde_kernelfile.json')
reco_service.store_kernels(outfile)
```
